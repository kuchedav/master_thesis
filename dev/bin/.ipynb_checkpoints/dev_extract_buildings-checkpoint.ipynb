```python
# import the necessary packages
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage import io
import matplotlib.pyplot as plt
import argparse

image = mapvault[0].copy()

segments = slic(image, n_segments = 3000, sigma = 1)

fig, axs = plt.subplots(1,1,figsize=(10, 10))
axs.imshow(mark_boundaries(image, segments, color=(1, 0, 0)))
```

```python
image_filtered = image.copy()

for i in range(0,np.max(segments)):
    image_filtered[segments==i] = 255 if np.mean(image[segments==i]) > 30 else 0
```

```python
fig, axs = plt.subplots(1,2,figsize=(20, 20))
axs[0].imshow(mark_boundaries(image_filtered, segments, color=(1, 0, 0)))
axs[1].imshow(image)
```

```python
fig, axs = plt.subplots(1,2,figsize=(20, 20))

# overlay the map
added_map = cv2.addWeighted(mapvault[0][:,:,0:3],0.2,image_filtered,0.8,0)
axs[0].imshow(added_map)

# overlay the image
added_image = cv2.addWeighted(imgvault[0][:,:,0:3],1,image_filtered,0.5,-80)
axs[1].imshow(added_image)
```
